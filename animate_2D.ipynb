```python
import sdf_helper as sh
from matplotlib.colors import SymLogNorm, LogNorm
from scipy.constants import elementary_charge
from metadata import get_simulation

# if you are using vscode, add repo directory path to python.analysis.extraPaths in settings.json to enable syntax highlighting for dynamic imports
from plot_helper_2d import animate_field
from utils import Scalar, Species, Vector
```

```python
simulation_id = '0.2_area_20240227'
simulation = get_simulation(simulation_id)
```

```python
ani, ax, cbar = animate_field(
    simulation.data_dir_path,
    Vector.Ey,
    normalization_factor=simulation.laser.normalized_amplitude,
    norm=SymLogNorm(linthresh=1e-2, linscale=1),
    cmap="bwr"
)

cbar.set_label(r"$\frac{eE}{m_e c\omega}$")
ax.set_title("Ey")
ax.set_xlabel("x [um]")
ax.set_ylabel("y [um]")

ani.save(simulation.analysis_dir_path + "/Ey_movie.mp4", writer="ffmpeg", fps=10, dpi=300)
```

```python
ani, ax, cbar = animate_field(
    simulation.data_dir_path,
    Scalar.CHARGE_DENSITY,
    species=Species.ALL,
    normalization_factor=simulation.laser.critical_density * elementary_charge,
    smoothing_sigma=3,
    norm=SymLogNorm(linthresh=1e-2, linscale=1),
    cmap="bwr"
)

cbar.set_label(r"$\frac{\rho}{n_c e}$")
ax.set_title("Charge density")
ax.set_xlabel("x [um]")
ax.set_ylabel("y [um]")

ani.save(simulation.analysis_dir_path + "/charge_density_movie.mp4", writer="ffmpeg", fps=10, dpi=300)
```

```python
from matplotlib.colors import LogNorm

ani, ax, cbar = animate_field(
    simulation.data_dir_path,
    Scalar.NUMBER_DENSITY,
    species=Species.ELECTRON,
    normalization_factor=simulation.laser.critical_density,
    norm=LogNorm(vmin=1e-5, vmax=2e1),
    cmap="inferno",
)

cbar.set_label(r"$\frac{n_e}{n_c}$")
ax.set_title("Electron Number Density")
ax.set_xlabel("x [um]")
ax.set_ylabel("y [um]")

ani.save(simulation.analysis_dir_path + "/electron_number_density_movie.mp4", writer="ffmpeg", fps=10, dpi=300)
```
