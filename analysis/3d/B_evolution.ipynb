```python
import numpy as np
import sys
import os 
sys.path.append(os.getenv("EPOCH_SIMS_HOME"))
from plot_helper import animate_field
from configs.config import *
%config InlineBackend.figure_format = 'retina' 
```

```python
# load grid data
grid = np.load(
    os.path.join(raw_data_folder, "grid.npy"), allow_pickle=True
)
x_array, y_array, z_array = grid
```

## Bz
Y is the polarization direction

### XY plane

```python
Bz_XY = np.load(
    os.path.join(raw_data_folder, "Magnetic_Field_Bz_XY.npy"),
    allow_pickle=True,
)
```

```python
Bz_XY.shape
```

```python
# call animate_density and save animation
extent = [x_array[0], x_array[-1], y_array[0], y_array[-1]]
ani, ax = animate_field(Bz_XY, extent, log_scale=False)

# set axes label
ax.set_xlabel("x [m]")
ax.set_ylabel("y [m]")
ax.set_title(f"Bz")
ani.save(os.path.join(media_folder, 'Bz_xy.mp4'), fps=10, dpi=300)
```

### YZ plane

```python
Bz_YZ = np.load(
    os.path.join(raw_data_folder, 'Magnetic_Field_Bz_YZ.npy'),
    allow_pickle=True,
)
```

```python
# call animate_density and save animation
extent = [y_array[0], y_array[-1], z_array[0], z_array[-1]]
ani, ax = animate_field(Bz_YZ, extent, log_scale=False)

# set axes label
ax.set_xlabel("y [m]")
ax.set_ylabel("z [m]")
ax.set_title(f"Bz")
ani.save(os.path.join(media_folder, f"Bz_yz_log.gif"), fps=10, dpi=300)
```
