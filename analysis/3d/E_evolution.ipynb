```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from plot_helper import animate_field
import sys 
sys.path.append('../..')
from configs.config import *
%config InlineBackend.figure_format = 'retina' # disable this if not using a retina display
```

```python
# load grid data
grid = np.load(os.path.join(raw_data_folder, "grid.npy"), allow_pickle=True)
x_array, y_array, z_array = grid
```

## Ey 
Y is the polarization direction

### XY plane

```python
Ey_XY = np.load(
    os.path.join(raw_data_folder, "Electric_Field_Ey_XY_0_321_10.npy"), allow_pickle=True
)
```

```python
Ey_XY.shape
```

```python
# call animate_density and save animation
extent = [x_array[0], x_array[-1], y_array[0], y_array[-1]]
ani, ax = animate_field(Ey_XY, extent, log_scale=False)

# set axes label
ax.set_xlabel("x [m]")
ax.set_ylabel("y [m]")
ax.set_title(f"Ey")
ani.save(os.path.join(media_folder, 'Ey_xy.gif'), fps=10, dpi=300)
```

### YZ plane

```python
Ey_YZ = np.load(
    os.path.join(raw_data_folder, "Electric_Field_Ey_YZ_0_321_10.npy"), allow_pickle=True
)
```

```python
# call animate_density and save animation
extent = [y_array[0], y_array[-1], z_array[0], z_array[-1]]
ani, ax = animate_field(Ey_YZ, extent, log_scale=False)

# set axes label
ax.set_xlabel("y [m]")
ax.set_ylabel("z [m]")
ax.set_title(f"Ey")
ani.save(os.path.join(media_folder, f"Ey_yz_log.gif"), fps=10, dpi=300)
```
